```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv("C:/Users/shevd/Downloads/Training files/stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
plt.figuresize = (12,6) 
#create color variable
colors = ["#011638", "#7e2787", "ef2026"]
#create lable variable
labels = ["Deisel", "Petrol", "Gas"]
# change the style
sns.set_style("white")
# create the staked area chart
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              colors = colors, # add the color 
              edgecolor = "none" # remove the edges in the chart
          )
plt.xticks(df_fuel_engine_types["Year"], rotation = 45) # show all the x axis values and rotate them
plt.legend(labels = labels, loc = "upper left") # add the labels and locate them
sns.despine() #remove the border from the chart after change the style
plt.ylabel("Number of Cars", fontsize = 13) #add y axis label 
plt.title("Pupularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, fontweight = "bold") #add Title
plt.show()
```
