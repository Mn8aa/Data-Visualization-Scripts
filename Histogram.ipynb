```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
```

```python
df_real_state = pd.read_csv("C:/Users/shevd/Downloads/Training files/Histogram_data.csv")
```

```python
sns.set_style("white") # chage the style of the figure to white backround 
plt.figure(figsize= (8,6)) # make the figure larger 
plt.hist(df_real_state["Price"],
        bins = 8,
        color = "#108A99") # create the histogram and set the nember of bins, and the color 
plt.xlabel("Price in (000'$)") # add a axis title
plt.ylabel("Number of Properities") # add y axis title
plt.title("Distribution of Real State Prices", fontsize = 14, weight= "bold") # add title to the figure 
sns.despine() # remove the top and the right borders from the figure
plt.show()
```
