```python
import pandas as pd
import matplotlib.pyplot as plt 

from matplotlib.ticker import PercentFormatter # to change the secondary y axis to percentages
```

```python
df_KDnuggets = pd.read_csv("C:/Users/shevd/Downloads/Training files/bar_line_chart_data.csv")
```

```python
fig, ax = plt.subplots(figsize = (10,8))

# first create the bar chart 
ax.bar(df_KDnuggets["Year"], # x axis
      df_KDnuggets["Participants"], # y axis
      color = "k") # set color to black
# format the bar chart 
ax.set_ylabel("Number of Participants",
             weight = "bold")
ax.tick_params(axis = "y",
              size = 2, 
              labelsize = "large")

#second create the line chart 
ax1 = ax.twinx() # plot the line in top the bar chart 
ax1.plot(df_KDnuggets['Year'], # x axis
       df_KDnuggets["Python Users"], # y axis
       color = "#b60000", # set color to dark red
       marker = "D") # set the marker type to dimond shape
# format the line chart 
ax1.set_ylabel("Python Users", #add y title
             weight = "bold")
ax1.tick_params(axis = "y", # format y axis
                size = 2, 
                colors = "#b60000",
                labelsize = "large")
ax1.set_ylim(0,1) # set limits to the y axis
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0)) # change the values to percentages
ax1.set_title("KD Nuggets Survey Python Users (2012 - 2019)" , fontsize = "14" , weight = "bold") # add title to the figure
```
