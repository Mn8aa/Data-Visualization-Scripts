```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("C:/Users/shevd/Downloads/Training files/line_chart_data.csv")
```

```python
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
```

```python
plt.figure(figsize = (20,8)) # increase the chart size

plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["GSPC500"]) # add the first line to the chart 
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["FTSE100"]) # add the second line 

#  add title and labels to the chart 
plt.title("S&P VS FTSE Returns (2000 - 2010)", fontsize = 14 , fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")

# add legend to the chart 
label = ["GSP500", "FTSE100"]
plt.legend(labels= label, fontsize = "large")
plt.show()
```

```python
# filter the data for only the two quarters of 2008
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') &
                                      (df_spx_ftse_00_10.new_date <= '2008-12-31')]
```

```python
# create the line chart with the filter data

plt.figure(figsize = (20,8)) # increase the chart size

plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["GSPC500"] , color = 'midnightblue') # add the first line to the chart 
plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["FTSE100"] , color = "crimson" ) # add the second line 

#  add title and labels to the chart 
plt.title("S&P VS FTSE Returns (H2 2008)", fontsize = 16 , fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels= label, fontsize = "large")

plt.show()
```
