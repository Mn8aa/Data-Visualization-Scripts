```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("C:/Users/shevd/Downloads/Training files/pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
sns.set_palette('colorblind') #change the color theme
```

```python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels = df_fuel_engine_type['Engine Fuel Type'].values, #create the pie chart with the data
       autopct = '%.2f%%', #make the percentge with two decimals
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold',
                   'rotation' : '30',
                   'color': 'w'} ) #edit the text in the pie chart
plt.legend()
plt.title('Cars by Fuel Engine Type',
         fontweight = 'bold',
         fontsize = 18) # add title to the pie chart 
plt.show()
```
