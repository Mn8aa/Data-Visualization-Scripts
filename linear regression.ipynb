```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_exenditure = pd.read_csv("C:/Users/shevd/Downloads/Training files/scatter_plot_ii.csv")
```

```python
sns.set(rc = {'figure.figsize': (9,6)}) #increase the figure size using seaborn 

sns.regplot(x = "Budget", # x axis
            y = "Sales", # y axis
            data = df_ad_exenditure, # the data
            scatter_kws = {'color': 'k'}, # change the color of the points
            line_kws = {'color': 'red'}) # change the color of the line
# add labels to the figure
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show()
```

```python
# sns.regplot() vs sns.lmplot()
# -- sns.regplot() accepts a varity of formats for the x and y variables such as numpy arrays, pandas series, etc.
# -- sns.lmplot() support a wide range of features 
```

```python
sns.lmplot(x = "Budget",
           y = "Sales",
           data = df_ad_exenditure,
           height = 10, # increase the figure size
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show()
```
