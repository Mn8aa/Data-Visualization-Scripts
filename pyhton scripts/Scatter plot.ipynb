```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_state = pd.read_csv("C:/Users/shevd/Downloads/Training files/Histogram_data.csv")
```

```python
plt.figure(figsize = (12,8)) # change the figure size
# create the scatter plot
scatter = plt.scatter(df_real_state["Area (ft.)"], # add x axis
           df_real_state["Price"], # add y axis
           alpha = 0.6, # set the transparency 
           c = df_real_state["Building Type"], # color the plots depending on anohter variable
           cmap = 'viridis') # choose the color theme
# create a legend for the figure
plt.legend(*scatter.legend_elements(), 
          loc = "upper left",
          title = "Building Type")
# create titles for the figure
plt.title("Relationship between Area and price of California Real Estate")
plt.xlabel("Area (sq.ft)" , fontweight = "bold")
plt.ylabel("Price (000's $)", fontweight = "bold")
plt.show()
```
